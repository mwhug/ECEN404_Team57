# Get water uptake totals for each nutrient solution

```python
waterdf.drop(columns = ['Date'], inplace=True)
waterdf.head(12)
```

```python
waterdf.info()
```

```python
total_uptakes = {'H+C': 0, 'H+PAW': 0, 'H+K (100 ppm)': 0, 'H+K (200 ppm)': 0, 'H+K (300 ppm)': 0, 'H+Mg (30 ppm)': 0, 'H+Mg (50 ppm)': 0, 'H+Mg (70 ppm)': 0,
                 'H+N (100 ppm)': 0, 'H+N (175 ppm)': 0, 'H+N (250 ppm)': 0}
sum = 0
for column in waterdf:
  for row in range(11):
    wateruptake = waterdf[column][row]
    sum = sum + wateruptake
  total_uptakes[column] = sum
  sum = 0

print(total_uptakes)
```

```python
comparator = -1
highest_uptake_solution = ''
for key in total_uptakes:
  if total_uptakes[key] > comparator:
    comparator = total_uptakes[key]
    highest_uptake_solution = key
```
